```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import colormaps as cmaps
import matplotlib

# Set font settings
matplotlib.rcParams['font.family'] = "sans-serif"
matplotlib.rcParams['font.sans-serif'] = "Arial"

# Read data
data = pd.read_csv('suppfig4.csv')

# Set colors for plotting
color1 = cmaps.lipari[135].colors
color2 = cmaps.lipari[75].colors
colors = [color1, color2]

# Create figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(4.06, 4.75), sharey=True, sharex=True)

# Plot JSD data
sns.stripplot(data=data, y="JSD", x='Type', ax=axs[0], hue='Type', palette=colors, legend=False)
sns.boxplot(data=data, y='JSD', x='Type', color='White', ax=axs[0])
axs[0].set_title('Jensen\nShannon', fontsize=14)
axs[0].set_xlabel('')
axs[0].set_ylabel('')
axs[0].set_xticks([0.0, 1.0], labels=['Essential', 'Nonessential'])

# Style JSD plot
ax = axs[0]
ax.xaxis.set_tick_params(width=2, labelsize=11)
ax.yaxis.set_tick_params(width=2, labelsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

# Plot Hellinger data
sns.stripplot(data=data, y="Hellinger", x='Type', ax=axs[1], hue='Type', palette=colors, legend=False)
sns.boxplot(data=data, y='Hellinger', x='Type', color='White', ax=axs[1])
axs[1].set_title('Hellinger\nDistance', fontsize=14)
axs[1].set_xlabel('')
axs[1].set_xticks([0.0, 1.0], labels=['Ess', 'Noness'])

# Style Hellinger plot
ax = axs[1]
ax.xaxis.set_tick_params(width=2, labelsize=11)
ax.yaxis.set_tick_params(width=0, labelsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_linewidth(1.5)

fig.tight_layout()
plt.show()
```
